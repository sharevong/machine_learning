### 描述性统计

*总体*是一个特定研究中所有感兴趣个体的集合  
*样本*是从一个总体中选择出来的个体的集合，通常在研究中被期望代表总体  
*参数*是一个值，通常是一个数字值，它描述了一个总体。参数可以从单个测量中得到，或从对总体的一组测量中推导出来  
*统计量*是一个值，通常是一个数字值，它描述了一个样本。统计量可以从单个测量中得到，或从对样本的一组测量中推导出来  
*数据*是测量或观察，它通常被称为一个分数或原始分数  
*描述性统计*是用于总结，组织或简化数据的统计过程  
*推论性统计*是允许我们研究样本，然后将研究结果推广到样本来自的总体的技术  
*变量*是一种针对不同个体具有不同值的特性或条件  
*常量*是一个特性或条件，它不会变化，并且对每一个个体都是相同的  
*集中趋势*是一个统计量，它确定了能够代表分布中心的一个数值。集中趋势的目的是找到最典型或最能代表整个组的单个数值。描述集中趋势的统计量有平均数、中位数、众数。  
*中位数*是恰好将一个分布一分为二的数值。分布中恰好有50%的个体具有小于或者等于中数的数值。中数也相当于第50个百分位数。 
*众数*是在一个频数分布中，具有最大频率的分数或类别。  
*偏态分布*是非对称分布，尾在右侧的偏态分布被称为*正偏态分布*，尾在左侧的偏态分布被称为*负偏态分布*。  
*变异性*提供了对一个分布中的数据分散开或聚集在一起的程度的数量测量。描述变异性的统计量有全距、四分位距、标准差和方差。
*全距*是在最大的X值上实限和最小的X值下实限之间的差值。  
*四分位距*是被分布的中间50%覆盖的距离：四分位距=Q3-Q1。  
*离差*是数据到平均数的距离：数据的离差=$X-\mu$  
*方差*是离差平方的平均数。
*标准差*是方差的开根号：标准差=$\sqrt{方差}$    
*SS*：离差平方和。$SS=\sum{(X - \mu)^2}$  
*自由度*：对于一个有n个数值的样本，样本方差的自由度或$df$，被定义为：$df = n - 1$。自由度决定了样本中独立的和可以自由改变的数值个数。
*有偏*与*无偏*：如果从很多不同样本中得到的样本统计量的平均数等于相应的总体参数值，则一个样本统计量是*无偏*的；如果一个样本统计量的平均数总是低估或高估相应的总体参数，那么这个统计量是*有偏*的。其中样本方差（计算时用了n-1）是对总体方差的一个精确的无偏估计。  

描述性统计常用的统计量：平均数、频数、频数分布表、百分数、四分位数、中位数、众数

总体平均数：$\mu = \frac{\sum X}{N}$  
样本平均数：$M = \frac{\sum X}{n}$  
总体方差：$\sigma^2 = \frac{SS}{N}$  
总体标准差：$\sigma = \sqrt{\frac{SS}{N}}$  
MAD（median absolute derivation）：$MAD = median(\mid X_i - median(X_i)\mid)$    
样本方差：$s^2 = \frac{SS}{n - 1} = \frac{SS}{df}$  
样本标准差：$s = \sqrt{\frac{SS}{n - 1}} = \sqrt{\frac{SS}{df}}$  
z分数：$z = \frac{X - \mu}{\sigma}$  
正态分布：$Y = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-(x-\mu)^2/2\sigma^2}$  


### 推论性统计

*z分数*：指出了每一个X值在分布中的精确位置。z分数的符号+或-表示这个分数比平均数高或者比平均数低。  
*二项分布*：伯努利实验重复n次进行，事件在样本中出现的次数分布。二项分布显示了从X=0到X=n的每个X值相应的概率  
*二项分布的正态近似*：当n较大时，更具体地说，当pn和qn都大于等于10的时候，二项分布将会是一个近似的正态分布，在这种情况下，二项分布将会是近似于平均数为$\mu=pn$标准差为$\sigma=\sqrt{npq}$的正态分布。  
*标准正态分布*：均值为0，标准差为1的正态分布。一般地，一个落在超过$z=\pm1.96$边界以外的样本极端地不可能出现，“极端地不可能”被定义为小于等于5%的概率。  
*假设检验*：是一种统计方法，它使用样本数据评估关于总体参数的假设。 在假设检验中，可能存在2种不同的误差，称为第一类误差和第二类误差。 
*虚无假设*：$H_0$陈述处理没有效应。  
*备择假设*：$H_1$陈述处理有效应，总体存在改变、差异。  
*$\alpha$水平*：显著水平，一个概率值，被用于定义在虚无假设是正确的情况下非常不可能出现的样本结果。  
*临界区域*：是由在虚无假设正确的情况下非常不能出现的极端样本值组成的临界区域界限，由$\alpha$水平决定。如果样本数据位于临界区域中，则拒绝虚无假设。  
*第一类误差*：发生在研究者拒绝虚无假设，但事实上它并没有效应。假设检验的$\alpha$水平是导致第一类误差的概率，即$\alpha$水平确定了在虚无假设正确的情况下得到样本数据在临界区域的概率。  
*第二类误差*：发生在研究者接受虚无假设，但事实上它确实产生了效应。意味着假设检验不能检测出真正的处理效应。  
*统计上显著的*：当虚无假设正确时，一个结果非常不可能出现。那么，这个结果就被称为是显著的或者统计上显著的。  
*科恩（Cohen）d值*：测量效应的大小。$科恩d值 = \frac{平均数差}{标准差}$  
*统计效能*：是检验能正确地拒绝错误的虚无假设的概率，即效能是检验能够识别真正存在的处理效应的概率。  
*样本误差*是样本数据和它对应着的总体参数之间的差异。
*样本平均数的分布*是在一个总体中所有可能的固定大小（n）的随机样本平均数的集合。  
*中心极限定理*：对于任意平均数为$\mu$、标准差为$\sigma$的总体，样本大小为n的样本平均数分布具有平均数$\mu$，标准差$\sigma/\sqrt{n}$，当n趋于无穷时，分布将趋于正态。  一般的当样本大小达到n=30时，分布就已经几乎是正态的。  
*M的期望值*：样本平均数分布的平均数。  
*M的标准误*：样本平均数分布的标准差。标准误估计了由于随机性造成的M和$\mu$之间的标准差量。  
*大数定理*：样本大小（n）越大，样本平均数同总体平均数接近的可能性越大。  

标准误：$\sigma_M = \frac{\sigma}{\sqrt{n}}$  
样本平均数的z分数：$z = \frac{M - \mu}{\sigma_M}$  

#### 案例1

假设心理学家在设计一个评估一种新生长激素效果的研究计划。已知正常的成年老鼠（没有激素）平均体重400克，标准差为20克，服从正态分布。心理学家计划选择一个由n=25只新生老鼠组成的样本，给他们注射激素，然后成年后测量它们的体重。怎么样确定激素是否有效果？

### t检验

*t分数*：当总体的标准差未知时，我们使用样本的标准差代替，这是z分数变成了t分数。  
*估计标准误*：当总体的标准差未知时，标准误也是未知的，我们用样本标准差来代替总体的标准差来计算标准误，它提供了一个从样本平均数M到其总体平均数$\mu$之间标准距离的估计。  
*t分布*：从总体里抽取的每一个样本都可以用来计算z分数和t分数。如果选出具有特定样本大小（n）的所有可能样本，得到的所有z分数形成一个z分数分布（z分布），与此相同，得到的所有t分数形成一个t分布。t分布接近正态分布，接近的程度由自由度决定，自由度越大，t分布越接近于正态分布。 
*方差百分率*：当清除处理效应时，变异性减少的程度，被称为方差百分率$r^2$。  一般当$r^2>0.25$时，认为由比较大的处理效应。  

估计标准误：$s_M = \sqrt{\frac{s^2}{n}}$  
t分数：$t = \frac{M - \mu}{s_M}$  
自由度：$df = n - 1$  
估计d值：$d = \frac{M - \mu}{s}$  
方差百分率：$r^2 = \frac{t^2}{t^2 + df}$

对于t分数的假设检验必需的2个基本假定为：
1. 样本中的数值必需包含互相独立的观察
2. 样本来自的总体必需是正态分布。但是当样本大小相对来说比较大时，违反这个假设对t分数所得到的结果的实际影响不大。  

#### 案例2

研究者想研究以蛾类为食的鸟类是否真的会对类似眼睛的图案表现出回避，进行了以下实验：选一组吃蛾子的鸟为样本，n=16，这些鸟放在通过由2个隔间组成的笼子里，鸟可以通过隔间之间的门在2个隔间中自由行动。其中一间的壁上有2个类似眼睛的圆点。研究者将每只鸟放在2个隔间的门道上，每只鸟都被放在这个装置中60分钟，记录它们在没有圆点的那间隔间里停留的时间。假设研究者手机样本数据时，发现那些鸟在没有圆点的隔间里待的时间时M=35分钟，那么是否能确定鸟类对类似眼睛的图案有回避？

### 独立测量t校验

*独立测量研究*：每个处理条件（或每个总体）都应用单独样本的研究设计。独立测量研究的目的是评价2个总体（或2个处理条件）之间的平均数差。  

全面的t分数公式：$t = \frac{(M_1 - M_2) - (\mu_1 - \mu_2)}{s_(M_1 - M_2)}$  
自由度：$df = {df}_1 + {df_2} = (n_1 - 1) + (n_2 - 1)$   
$M_1 - M_2$估计标准误：$s_(M_1 - M_2) = \sqrt{\frac{s^2_p}{n_1}+\frac{s^2_p}{n_2}}$  
合并方差：$s^2_p = \frac{SS_1 + SS_2}{df_1 + df_2}$  
估计d值：$d = \frac{M_1 - M_2}{\sqrt{s^2_p}}$  
方差百分率：$r^2 = \frac{t^2}{t^2 + df}$  


对了独立测量t校验，需要满足下面3个假设：
1. 每个样本的观察都是独立的；
2. 与2个样本相对应的2个总体都必须是正态分布的； 
3. 与2个样本相对应的2个总体都必须有相等的方差。这个假设也叫方差齐性。可以使用Hartley实验，来检验2个样本总体是否满足方差齐性。  

#### 案例3

一位发展心理学家想检查6岁女孩与6岁男孩在预研能力上的差异。样本是n=5的女孩和n=5的男孩，每个孩子都参加标准的语言能力测试。结果女孩组平均测试成绩M=83，离差平方和SS=200；男孩组M=71，SS=120，这些数据能说明6岁女孩与6岁男孩在语言能力上有差异吗？

### 相关测量t校验

*重复测量研究*：样本中的个体要在同一因变量下被测量多于一次，所有处理条件都使用相同的测试。  
*差值*：相关个体分数的差值。$D = X_2 - X_1$  

相关t分数：$t = \frac{M_D - \mu_D}{s_{M_D}}$  
自由度：$df = n - 1$，其中$n$为D值个数  
D值估计标准误：$s_{M_D} = \sqrt{\frac{s^2_D}{n}}$  
D值样本标准差：$s_D = \frac{SS_D}{df}$  
估计d值：$d = \frac{M_D}{s_D}$  
方差百分率：$r^2 = \frac{t^2}{t^2 + df}$

### 估计

*估计*：用样本统计量推断总体参数的推论性过程。  
*点估计*：使用单一的数值作为对未知量的估计。  
*区间估计*：使用值域对未知量进行估计。区间估计伴随着一定水平的信心（或可能性），称为置信区间。  

单一样本t分数的估计公式：$\mu = M \pm ts_M$  
独立测量t分数的估计公式：$\mu_1 - \mu_2 = M_1 - M_2 \pm ts_{M_1 - M_2}$  
重复测量t分数的估计公式：$\mu_D = M_D \pm ts_{M_D}$  

#### 案例5

研究者需要确定一个学区中三年级学生的平均阅读水平。样本为n=25个学生，每个学生接受一个标准化阅读成绩测验。样本平均数是M=72，离差平方和SS=2400。使用样本结果构建对于总体平均数99%的置信区间。

### 方差分析

*方差分析*：ANOVA是一个假设检验过程，用于评估两个或多个处理（总体）平均数的差异，与t检验相比，可以用来比较2个或更多的处理。与t校验类似，也分为独立测量方差分析和重复测量方差分析。  
*因素*：在方差分析中，用来区分比较组的变量称为因素。  
*水平*：组成一个因素的各个条件或是数值被称为这个因素的水平。  
*F分数*：ANOVA的检验统计量，$F=\frac{处理间方差}{处理内方差}$。  
*F分布*：F分布形状取决于F分数中2个方差的自由度，对于非常大的df值，所有的F值将集中在1.00左右，对于小一些的df值，F分布会更加分散。  


#### 独立测量方差分析

总的离差平方和：$SS_总 = \sum{X^2} - \frac{G^2}{N}$，其中$G = \sum{X}$  
处理间离差平方和：$SS_间 = SS_总 - SS_内 = \sum{\frac{T^2}{n}} - \frac{G^2}{N}$，其中$T$为每个处理条件的总和  
处理内离差平方和：$SS_内 = \sum{SS_{每个处理内}}$  
总的自由度：$df_总 = N - 1$  
处理间自由度：$df_间 = k - 1$，其中k为处理的个数  
处理内自由度：$df_内 = N - k$  
处理间方差：$MS_间 = \frac{SS_间}{df_间}$  
处理内方差：$MS_内 = \frac{SS_内}{df_内}$  
F分数：$F = \frac{MS_间}{MS_内}$  
方差百分率：$r^2 = \frac{SS_间}{SS_总}$  

#### 重复测量方差分析

重复测量方差分析相对于独立测量方差分析，在计算F分数上，需要将分母处理内方差包含的因个体差异引起的方差清除掉，因为由于重复测量，F分数的分子处理间方差已经消除了个体差异。

被试间离差平方和：$SS_{被试间} = \sum{\frac{P^2}{k}} - \frac{G^2}{N}，其中P为研究中每个个体的所有数据$  
被试间自由度：$df_{被试间} = n - 1$，其中$n$为被试个数。  
消除个体差异的处理内离差平方和：$SS_{误差} = SS_{处理内} - SS_{被试间}$  
清除个体差异的处理内自由度：$df_{误差} = df_{处理内} - df_{被试间}$  
清除个体差异的处理内方差：$MS_{误差} = \frac{SS_{误差}}{df_{误差}} = \frac{SS_{处理内} - SS_{被试间}}{df_{处理内} - df_{被试间}}$  
F分数：$F = \frac{MS_{处理间}}{MS_{误差}}$  
方差百分率：$\eta^2 = \frac{SS_{处理间}}{SS_{处理间} + SS_{误差}}$

### 卡方检验

*拟合度的卡方检验*：从样本数据中得到的比例来检验关于相应的总体比例的假设。检验确定了得到的样本比例与虚无假设中陈述的总体比例之间的拟合度。 
*观察频数*：样本中被归入特定类别的个体个数。每个个体都被归入一个类别，且只被归入一个类别。  
*期望频数*：定义了在样本比例与虚无假设中陈述的比例完美吻合的情况下产生的理想的、假设的样本分布。  
*卡方*：测量了数据$f_0$与假设$f_e$的拟合程度。  
*卡方分布*：正偏态的，卡方分布的精确形状由类别的个数确定。  

卡方：$\chi^2 = \sum{\frac{(f_o - f_e)^2}{f_e}}$  
卡方自由度：$df = C - 1$  
衡量效应大小：$\varphi = \sqrt{\frac{\chi^2}{n}}$，其中n为数据点个数


*独立性卡方检验*：可以用来检验2个变量之间是否存在关系。

期望频数：$f_e = \frac{f_cf_r}{n}$，其中$f_c$是列的频数总和，$f_r$为行的频数总和，$n$为数据点个数  
独立性卡方检验的自由度：$df = (R - 1)(C - 1)$

### FDR控制过程

*FDR*：false discovery rate。所谓discovery，是指拒绝虚无假设，叫有discovery。FDR衡量同时做多个假设实验时，出现错误拒绝虚无假设的概率。FDR通过Benjamini–Hochberg过程，控制整体虚无假设出现错误拒绝虚无假设的概率。

当同时做多个虚无假设时，推断可能出现以下结果：

||Null hypothesis is true($H_0$)|Alternative hypothesis is true($H_A$|Total|
|---|---|---|---|
|Test is declared significant|V|S|R|
|Test is declared non-significant|U|T|m-R|
|Total|$m_0$|$m-m_0$|m|

其中m为假设检验个数，$m_0$为虚无假设为真的个数，未知参数，$m-m_0$为备择假设为真的个数，V为第一类误差的个数（false discovery），S为true discoveris个数，T为第二类误差个数，U为true negavtives个数。

$FDR = Q_e = E[Q] = E[\frac{V}{V+S}]$  

#### Benjamini-Hochberg过程

控制FDR在$\alpha$水平以下，过程如下：

1. 首先计算$H_1$...$H_m$ m个假设检验的p-value，表示为$P_1$...$P_m$，然后按照升序排列，记为$P_{(1)}$...$P_{(2)}$  
2. 对于给定$\alpha$，找到最大k满足$P_{k}\leq\frac{k}{m}$  
3. 拒绝掉i=1...k的虚无假设

通过BH过程，能够保证：

$FDR = E[Q] \leq \frac{m_0}{m}\alpha \leq \alpha$  

### Family-wise error rate

*FWR*: family-wise error rate。衡量同时做多个假设实验时，至少出现一次错误拒绝虚无假设的概率。与FDR相比，对第一类误差控制较强。
